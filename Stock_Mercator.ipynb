```python
!pip install pandas-datareader
import numpy as np
import pandas as pd
import pandas_datareader as pdr
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pdr.get_data_yahoo('MRC.WA','2020-04-08')
```

```python
df.head()
```

```python
df.tail()
```

```python
df['Close'].plot()
```

```python
df["MA_30"]=df['Close'].rolling(30).mean()
df["MA_60"]=df['Close'].rolling(60).mean()
df["MA_200"]=df['Close'].rolling(200).mean()
```

```python
df.describe()
```

```python
plt.figure(figsize=(20,10))
plt.plot(df['Close'], label='MERCATOR')
plt.plot(df['MA_30'], label='MERCATOR_30')
plt.plot(df['MA_60'], label='MERCATOR_60')
plt.plot(df['MA_200'], label='MERCATOR_200')
plt.legend(loc=2)
```
